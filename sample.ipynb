```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Example data: Date, Time, Altitude
data = {
    'Date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04'],
    'Time': ['00:00', '01:00', '02:00', '03:00'],
    'Altitude': [20, 35, 50, 60]
}

df = pd.DataFrame(data)

# Parameters for the trigonometric function
A = 20 # Amplitude
D = 40 # Vertical Shift
B = 2 * np.pi / 30 # Frequency (assuming a period of 30 days)
C = 1 # Phase Shift (example value)

# Generate time values
times = np.arange(0, 4) # Example time range in hours
model_altitude = A * np.sin(B * (times - C)) + D

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(times, model_altitude, label='Modelled Altitude', marker='o')
plt.scatter(times, df['Altitude'], color='red', label='Actual Data')
plt.xlabel('Time (hours)')
plt.ylabel('Altitude (degrees)')
plt.title('Celestial Body Altitude Model')
plt.legend()
plt.grid(True)
plt.show()
```
